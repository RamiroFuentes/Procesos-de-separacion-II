```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.arange(.0,1.05,.05)
alpha = 2.5
y = (x*alpha)/(1+(alpha-1)*x)

data = pd.DataFrame({"X":x,"Y":y})

display(data)
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)

fig, ax = plt.subplots(figsize=(10, 8),dpi=200)

# Set axis ranges; by default this will put major ticks every 1 .
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)

# Change major ticks to show every.1 .
ax.xaxis.set_major_locator(MultipleLocator(.1))
ax.yaxis.set_major_locator(MultipleLocator(.1))

# Change minor ticks to show every 10.
ax.xaxis.set_minor_locator(AutoMinorLocator(10))
ax.yaxis.set_minor_locator(AutoMinorLocator(10))

# Turn grid on for both major and minor ticks and style minor slightly
# differently.
ax.grid(which='major', color='#CCCCCC', linestyle='--')
ax.grid(which='minor', color='#CCCCCC', linestyle=':')

ax.plot(x,y,marker=".")
ax.plot([x[0],x[-1]],[y[0],y[-1]])
ax.plot([.55,.0],[.55,.11])



plt.title("Mezcla Benceno-Tolueno")
plt.xlabel("X")
plt.ylabel("Y")

plt.grid()

plt.show()
```
