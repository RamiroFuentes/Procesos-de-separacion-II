## Gráfica de T de rocío 

$f(\psi)=\sum_{i=1}^{c}\frac{z_i[1-k_i]}{1+\psi[k_i-1]}$

Fracciones del problema realizado en clase:

$z_{n-Butano}=0.2$

$z_{n-Pentano}=0.25$

$z_{n-Hexano}=0.25$

$z_{n-Heptano}=0.3$

Constantes del problema realizado en clase:


$K_{n-Butano}=4$

$K_{n-Pentano}=1.9$

$K_{n-Hexano}=1$

$K_{n-Heptano}=0.5$

Desarrollando la expresión:

$f(\psi)=\frac{z_1[1-k_1]}{1+\psi[k_1-1]} + \frac{z_2[1-k_2]}{1+\psi[k_2-1]} + \frac{z_3[1-k_3]}{1+\psi[k_3-1]} + \frac{z_4[1-k_4]}{1+\psi[k_4-1]}$

```python
# Librerias 
import numpy as np
from numpy import poly1d,polyfit  
import matplotlib.pyplot as plt
from sympy import Symbol

# Para imprimir en formato LaTex
from sympy.interactive import printing
printing.init_printing(use_latex=True)
```

## Derivada de la función


```python
psi = Symbol('psi')

z_1 = Symbol('z_1')
z_2 = Symbol('z_2')
z_3 = Symbol('z_3')
z_4 = Symbol('z_4')

k_1 = Symbol('k_1')
k_2 = Symbol('k_2')
k_3 = Symbol('k_3')
k_4 = Symbol('k_4')

f_psi= ((z_1*(1-k_1)/(1+psi*(k_1-1))) + (z_2*(1-k_2)/(1+psi*(k_2-1))) + (z_3*(1-k_3)/(1+psi*(k_3-1))) + (z_4*(1-k_4)/(1+psi*(k_4-1))))
dpsi = f_psi.diff(psi)

print('Función:')
display(f_psi)

print('Derivada:')
display(dpsi)
```

## Funciones propuestas

```python
def Rachford_Rice_4(z,k,Li,Ls,p):
    psi = np.arange(Li,Ls+p,p)
    f_psi = []
    for value in psi:
        f_psi.append((z[0]*(1-k[0])/(1+value*(k[0]-1))) + (z[1]*(1-k[1])/(1+value*(k[1]-1))) + (z[2]*(1-k[2])/(1+value*(k[2]-1))) + (z[3]*(1-k[3])/(1+value*(k[3]-1))) )
        
    return psi,f_psi

def find_root_interval(given):
    i = 0
    for value in given:
        if value >=0:
            P_1 = i
            P_0 = i-1
        else: 
            i += 1
    return P_0,P_1

def interp(x_1,x_2,y_1,y_2,y_n):
    x_n = x_1 -((x_1-x_2)*(y_1-y_n)/(y_1-y_2))
    return x_n , y_n
```
