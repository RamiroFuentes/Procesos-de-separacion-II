## Derivada de la función

```python
psi = Symbol('psi')

z_1 = Symbol('z_1')
z_2 = Symbol('z_2')
z_3 = Symbol('z_3')
z_4 = Symbol('z_4')

k_1 = Symbol('k_1')
k_2 = Symbol('k_2')
k_3 = Symbol('k_3')
k_4 = Symbol('k_4')

f_psi= ((z_1*(1-k_1)/(1+psi*(k_1-1))) + (z_2*(1-k_2)/(1+psi*(k_2-1))) + (z_3*(1-k_3)/(1+psi*(k_3-1))) + (z_4*(1-k_4)/(1+psi*(k_4-1))))
dpsi = f_psi.diff(psi)

print('Función:')
display(f_psi)

print('Derivada:')
display(dpsi)
```
