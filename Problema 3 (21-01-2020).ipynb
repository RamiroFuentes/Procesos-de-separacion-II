## Primer intento

```python
# Dise√±o de la gr√°fica por forma vectorizada considerando tres mezclas
psi = np.arange(0,1.1,.1)
k =[2.5,1.1,0.53] 
z = [.25,.4,.35]
f_psia= ((z[0]*(1-k[0])/(1+psi*(k[0]-1))) + (z[1]*(1-k[1])/(1+psi*(k[1]-1))) + (z[2]*(1-k[2])/(1+psi*(k[2]-1))))

# Obtencion del polinomio
Poly = poly1d(polyfit(psi,f_psia, deg=3), variable='x') 
print(Poly)

# Evaluacion del polinomio en el rango de nuestra psi
y = Poly(psi)

# Encontrando las raices de nuestro polinomio
[min,max]=find_root_interval(f_psia)
[x_n,y_n] = interp(psi[min],psi[max],f_psia[min],f_psia[max],0)
print(f"La Ra√≠z es:[{x_n},{y_n}]")

# Dimensiones de la grafica
plt.figure(figsize=(15,10))

# Grafica del polinomio
plt.plot(psi,y,marker=".",label=f"{Poly}")
plt.grid()

# Grafica para los valores de psi y F(psi)
plt.plot(psi,f_psia,marker='*',color="#00C72E",label="Funci√≥n sumatoria")

# Muestra del punto donde y = 0
plt.plot(x_n,y_n,marker="X",color="red",label=f"Ra√≠z = {x_n}")

# Prueba con el resultado de la sumatoria 
plt.axhline(y=0.0,color='#434F66')

# Lineas Verticales y horizontales
plt.axvline(x=1,color='r',ls='dotted')
plt.axvline(x=0,color='r',ls='dotted')
plt.axvline(x=x_n,color='b',ls='dotted')

# Formato de gr√°fica
plt.title("$f (\psi)$ vs $\psi$",fontsize=20)
plt.grid(True)
plt.xlabel('$\psi$', fontsize=15)
plt.ylabel('$f(\psi$)', fontsize=15).set_rotation(0)
plt.legend()
plt.show()
```
