## Problema 2

```python
C = [0.1,0.2,0.7]
K = [2.7,0.95,0.34]

# Para calcular los valores de nuestra P

def pressure_given_psi(P_interval,z,k,psi):
    Pressure = []
    for P in P_interval:
        F_psi = 0
        print(f"Presi√≥n = {P}")
        
        for i in np.arange(0,len(C)):
            Fact = (k[i]*200/P)
            F_psi = F_psi + z[i]*(1-Fact)/(1+(psi*(Fact-1)))
            print(F_psi)
        print()
        
        Pressure.append(F_psi)
        
    return Pressure

interval = np.arange(100,127,.1)
P = pressure_given_psi(interval,C,K,.1)
# print(P)
plt.plot(interval,P,marker="")
plt.grid()
```
